# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
data=pd.read_csv("C:/Users/nishn/Downloads/Ice Cream Sales - temperatures.csv")
data.head()
```

# Extracting X and y datasets

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

# Splitting into testing and training set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3,random_state=42)
```

# Training on training set using Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

# Prediction on Test Set

```python
y_pred=regressor.predict(X_test)
```

# Visualizing the Training Set

```python
plt.scatter(X_train,y_train,color="red")
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title("Temperature v/s Ice Cream Profits")
plt.xlabel("Temperature")
plt.ylabel("Ice Cream Profits")
plt.show()
```

# Visualizing the Test Set

```python
plt.scatter(X_test,y_test,color="red")
plt.plot(X_test,y_pred,color='blue')
plt.title("Temperature v/s Ice Cream Profits")
plt.xlabel("Temperature")
plt.ylabel("Ice Cream Profits")
plt.show()
```
